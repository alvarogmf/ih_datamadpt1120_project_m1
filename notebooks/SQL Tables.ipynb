```python
table_query = """SELECT country_info.uuid, country_info.country_code, 
career_info.normalized_job_code as parent_uuid, 
personal_info.gender AS 'Gender',
count(country_info.uuid) AS 'Quantity'
FROM country_info
JOIN career_info ON career_info.uuid = country_info.uuid
JOIN personal_info ON personal_info.uuid = country_info.uuid
GROUP BY country_info.country_code"""

main_table = pd.read_sql_query(table_query,engine)

main_table.head()
```

```python
import json
import requests
 
response = requests.get('http://api.dataatwork.org/v1/jobs')
results = response.json()
results
```

```python
jobs = pd.DataFrame(results)
jobs
```

```python
clean_jobs = jobs[['parent_uuid','title']]
clean_jobs
```

```python
final = pd.merge(main_table, clean_jobs, on="parent_uuid")
final
```
